```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

data = pd.read_csv('/content/sample_data/Lesson_4_student_grades_dataset.csv')
data.columns
```

```python
x = data['Previous_Exam_Scores']
y = data['Grades']

plt.scatter(x,y, color = 'green')
plt.xlabel('Independent Variable (Previous_Exam_Scores)')
plt.ylabel('Dependent Variable (Grades)')
plt.title('Scatter Plot of Data')
plt.show()
```

```python
x = data[['Previous_Exam_Scores']]
y = data[['Grades']]

model = LinearRegression()

model.fit(x,y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_

print('Slope: ', slope)
print('Intercept: ', intercept)
```

```python
plt.scatter(x,y, color = 'orange')
plt.plot(x,model.predict(x), color = 'green')
plt.xlabel('Independent Variable (Previous_Exam_Scores)')
plt.ylabel('Dependent Variable (Grades)')
plt.title('Scatter Plot of Data')
plt.show()
```

```python
new_x = [[np.random.randint(100,200)]]

predicted_y = model.predict(new_x)

print(f'Value of X: {new_x}')
print(f'Predicted Value: {predicted_y}')
```
