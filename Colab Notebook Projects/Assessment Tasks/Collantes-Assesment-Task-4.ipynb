# 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

## 2. Creating Synthetic Dataset

```python
data = pd.read_csv('/content/sample_data/Lesson_4_student_grades_dataset.csv')

data.columns
```

# 3. Converting to Dataframe

```python
data = pd.DataFrame(data,columns=['Hours_of_Study', 'Attendance_Percentage', 'Previous_Exam_Scores',
       'Extracurricular_Activities', 'Travel_Time_to_School'])

print(data.head(10))
```

# 4. Preparing Data for Modeling

```python
x = data.drop('Grades', axis=1)
y = data['Grades']
```

# 5. Train and Split Data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

# 6. Fitting the Linear Regression Model

```python
model = LinearRegression()

model.fit(x_train, y_train)
```

# 7. Making Predictions

```python
new_data = np.array([[7, 96, 95, 1, 59]])

y_predict = model.predict(x_test)
prediction = model.predict(new_data)

grade = (prediction/60)*100
```

# 8. Print Result

```python
mse = mean_squared_error(y_test,y_predict)

print('MSE: ',mse)
print(f'Predicted Grade: {grade[0]}')
```
