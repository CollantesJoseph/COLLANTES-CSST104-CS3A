1. Working Libraries


```python
import pandas as pd #Data Manipulation
import matplotlib.pyplot as plt #Data Visualization
```

2. Loading Datasets

```python
data = pd.read_csv('/content/sample_data/sales_data_2.csv', encoding = 'latin-1')
```

3. Data Loading and Initial Analysis

```python
statistics = data.describe()

statistics
```

4. Data Manipulation

```python
product_revenue = data.groupby('Product')['Revenue'].sum()

highest_revenue_product = product_revenue.idxmax()

average_quantity_per_day = data['Quantity Sold'].mean()
```

5. Data Visualization

```python
product_revenue.plot(kind='bar')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.title('Revenue by Product')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show
```

6. Analysis Report


```python
report = f"""
Analysis Report

1. Data Loading and Initial Analysis:
  Basic statistics of the dataset:
{statistics}
2. Data Manipulation:
  Total revenue for each product:
  {product_revenue}
  Product with the highest revenue:
  Average quantity sold per day: {average_quantity_per_day : .2f}
3. Data Visualization:
  Bar chart displaying revenue for each product is attached.
4. Insights:
  -The product {highest_revenue_product} generates the highest revenue.
  -On average, {average_quantity_per_day : .2f} units sold per day.

"""

print(report)
```

7. Saving the Report to a File

```python
with open('report.txt', 'w') as f:
  f.write(report)
```
