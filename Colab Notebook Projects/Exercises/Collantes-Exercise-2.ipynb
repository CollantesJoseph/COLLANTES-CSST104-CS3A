1. Creating Dataset

```python
import pandas as pd

data = {
    'student_id': range(1,21),
    'gender': ['Male','Female','Female','Female','Male','Female','Male',
               'Female','Female','Female','Male','Female','Male','Female',
               'Female','Female', '', 'Female', '', 'Female'],
    'test_date': pd.date_range(start = '2024-02-21', periods = 20),
    'study_hours': [2,4,5,2,5,7,8,1,3,4,2,5,3,2,8,5,4,5,3,6],
    'test_score': [88,89,80,85,95,77,88,91,83,94,92,95,83,82,96,85,84,85,83,96],
    'attendance': ['Present', '','Absent','Present', 'Absent', 'Present', 'Absent',
                   'Absent', 'Absent', '', 'Present', 'Absent', 'Absent', 'Absent',
                   'Absent', 'Absent', '', '', 'Present', '' ]
}

df = pd.DataFrame(data)

df.to_csv('student_performance.csv', index = False)
```

2. Data Loading and Initial Examination

```python
df = pd.read_csv('student_performance.csv')

#Display
print(df.head(15))

#Check for missing value
print(df.isnull().sum())
```

3. Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sne

#Compute summary statistics
summary_stats = df[['test_score', 'study_hours']].describe()

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sne.histplot(df['test_score'], bins=15, kde=True)
plt.title('Distribution of Test Score')

plt.subplot(1,3,2)
sne.histplot(df['study_hours'], bins=15, kde=True)
plt.title('Distribution of Study Hours')

plt.subplot(1,3,3)
sne.countplot(data=df, x='attendance')
plt.title('Distribution of Attendance')




```

4. Data Visualization

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sne.countplot(data=df, x='gender')
plt.title('Gender Distribution')

correlation_matrix = df[['study_hours', 'test_score', 'attendance']].corr()

plt.subplot(1,3,2)
sne.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Heat Map')


plt.figure(figsize=(11,4))
sne.lineplot(data=df, x='test_date', y='test_score')
plt.xticks(rotation=80)
plt.title('Trend in Test SCore Over Time')
```
