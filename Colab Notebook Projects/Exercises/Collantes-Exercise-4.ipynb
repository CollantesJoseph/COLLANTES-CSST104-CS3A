# 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# 2. Creating Synthetic Dataset

```python
np.random.seed(0)
x = np.random.rand(100,3)
y = x[:,0] * 150000 + x[:,1] * 100000 + x[:,2] * 50000 + 5000
```

# 3. Converting to Dataframe

```python
data = pd.DataFrame(x,columns=['Size', 'Bedroom', 'Age'])

data['Price'] = y
print(data.head())
```

# 4. Preparing Data for Modeling

```python
x = data[['Size', 'Bedroom', 'Age']]
y = data['Price']
```

# 5. Train and Split Data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

# 6. Fitting the Linear Regression Model

```python
model = LinearRegression()

model.fit(x_train, y_train)
```

# 7. Getting Model Coefficients

```python
y_predict = model.predict(x_test)
```

# 8. Mean Squared Error

```python
mse = mean_squared_error(y_test,y_predict)

print('MSE',mse)
print(y_predict)
```
